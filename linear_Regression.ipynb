```python
import pandas as pd
```

```python
frame = pd.read_csv("Data\data.csv")
xTrain = frame["x"].to_numpy(dtype=float)
yTrain = frame["y"].to_numpy(dtype=float)
xTrain
```

```python
def calculateYHat(x , i , w , b):
    yHat = w * x[i] + b
    return yHat
```

```python
def calculateCostFunc(x ,  y , w , b ):
    
    m = x.shape[0]
    cost = 0
    for i in range(0, m):
        cost += ((calculateYHat(x , i , w , b)  - y[i]) ** 2)
    cost /= (2 * m)
    
    return cost
    
```

```python
def calculate_gradient(x , y ,w ,b ):
    
    
    w_descent = 0 
    b_descent = 0 
    
    
    m = x.shape[0]
    
    for i in range(m):
        
        w_descent += (calculateYHat(x , i, w , b) - y[i]) * x[i]
        b_descent += (calculateYHat(x , i , w , b) - y[i])
    
    w_descent /= m
    b_descent /= m
    
    
    
    return w_descent , b_descent
```

```python
def gradient_descend(x ,y ,w , b , iteration_Num  , alpha):
    
    for item in range(iteration_Num):
        temp_W , temp_B = calculate_gradient(x , y ,w ,b)
        w = w - (alpha * temp_W)
        b = b - (alpha * temp_B)
    
    return w , b 
    print(w , b) 
```
